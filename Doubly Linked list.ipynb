```python
class Node:
    def __init__(self, data=None, next=None, prev=None):
        self.data = data
        self.next = next
        self.prev = prev
```

```python
class DoublyLinkedList:
    def __init__(self):
        self.head = None
        
    def insert_at_beginning(self,data):
        node = Node(data,self.head,None)
        if self.head == None:
            self.head = node
        else:
            self.head.prev = node
            self.head = node 
        
    def insert_at_end(self,data):
        if self.head is None:
            self.head = Node(data,None,None)
            return
        
        itr = self.head
        while(itr.next):
            itr = itr.next
        itr.next = Node(data,None,itr)
        return
    
    def insert_values(self,data_list):
        #self.head = None
        for data in data_list:
            self.insert_at_end(data)
            
    def get_last_node(self):
        itr = self.head
        while itr.next:
            itr = itr.next
        return itr
        
    def print_forward(self):
        if(self.head is None):
            print("Linked list is empty")
        else:
            itr = self.head
            lst1 = ''
            while itr:
                lst1 = lst1 + str(itr.data) + '-->'
                itr = itr.next
            print(lst1)

    def print_backward(self):
        if(self.head is None):
            print("Linked list is empty")
        else:
            last_node = self.get_last_node()
            itr = last_node
            lst1 = ''
            while itr:
                lst1 = lst1 + str(itr.data) + '-->'
                itr = itr.prev
            print("Reverse order:",lst1)

    def get_length(self):
        count = 0
        itr = self.head
        while itr:
            count = count+1
            itr = itr.next
        return count
                
    def insert_at(self,index,data):
        if index<0 or index>self.get_length():
            raise Exception("Invalid Index")
       
        if(index == 0):
            self.insert_at_beginning(data)
            return
        
        count = 0
        itr = self.head
        while itr:
            if(count==index-1):
                node = Node(data, itr.next, itr)
                if node.next:
                    node.next.prev = node
                itr.next = node
                break
            itr = itr.next
            count = count + 1
            
    def insert_before(self,value,data):
        itr = self.head
        if(itr.data==value):
            self.insert_at_beginning(data)
            return
    
        while(itr):
            if(itr.data==value):
                node = Node(data,itr1.next,itr1)
                itr1.next = node
                itr.prev = node
                break
            itr1 = itr
            itr = itr.next
        
    def insert_after(self,value,data):
        itr = self.head
        while(itr):
            if(itr.data==value):
                node = Node(data,itr.next,itr)
                if node.next:
                    node.next.prev = node
                itr.next = node
                break
            itr = itr.next
            
    def remove_at(self,index):
        if index<0 or index>self.get_length():
            raise Exception("Invalid Index")
        
        if(index==0):
            self.head = self.head.next
            self.head.prev = None
            return
        
        count = 0
        itr = self.head
        while itr:
            if(count == index-1):
                itr.next = itr.next.next
                itr.next.prev = itr
                break
            itr = itr.next
            count = count+1
            
    def remove_by_value(self,data):
        itr = self.head
        if(itr.data==data):
            self.head = itr.next
            self.head.prev = None
            return
        while(itr):
            if(itr.data == data):
                itr1.next = itr.next
                itr.next.prev = itr1
                break
            itr1 = itr
            itr = itr.next
            
    def remove_before(self,data):
        itr = self.head
        if(itr.data == data):
            print("Already the first index, no element is present before it")
            return
        if(itr.next.data == data):
            self.head = itr.next
            self.head.prev = None
            return
        while(itr):
            if(itr.next.data == data):
                itr1.next = itr.next
                itr.next.prev = itr1
                break
            itr1 = itr
            itr = itr.next


    def remove_after(self,data):
        itr = self.head
        count = 1
        length = self.get_length()
        while(itr):
            if(itr.data == data):
                if(count == length):
                    print("Reached the last node, no value is present after it")
                    return
                if(count == length-1):
                    itr.next = None
                    return
                itr.next = itr.next.next
                itr.next.prev = itr
                #itr.next.next = None
                break
            itr = itr.next
            count = count+1
        if(count > length):
            print("Value is not present")
            return
```
